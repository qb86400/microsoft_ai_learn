### 삼각함수

```python
# 참고: https://mathbang.net/496#gsc.tab=0 호도법과 라디안

angle = 90
radian = math.radians(angle)  # 각도를 라디안으로 변환
sin_value = math.sin(radian)
cos_value = math.cos(radian)
tan_value = math.tan(radian)

print(f"각도: {angle}도")
print(f"사인값: {sin_value}")
print(f"코사인값: {cos_value}")
print(f"탄젠트값: {tan_value}")
```

```python
### 소수점 이하 올림, 내림
num = 1299.5
ceil_value = math.ceil(num)    # 올림
floor_value = math.floor(num)  # 내림
round_value = round(num)       # 반올림

print(f"{num}의 올림: {ceil_value}")
print(f"{num}의 내림: {floor_value}")
print(f"{num}의 반올림: {round_value}")
```
